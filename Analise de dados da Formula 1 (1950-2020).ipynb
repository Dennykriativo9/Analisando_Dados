```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
# Carregando as base de dados 

circuito =pd.read_csv("circuits.csv",sep =";") # base de dados sobre o circuito
construtores =pd.read_csv("constructors.csv") # base de dados sobre o construtores
pilotos =pd.read_csv("drivers.csv",sep =";") # base de dados sobre os pilotos 
tempo_volta =pd.read_csv("lap_times.csv",sep =",")  # basse de dadso tempo de voltas
pit_stop =pd.read_csv("pit_stops.csv",sep =",") # base de dados sobre o Pit Stop
qualificacao =pd.read_csv("qualifying.csv",sep=",") # base de dados sobre a qualificação
corrida =pd.read_csv("races.csv",sep =";")  # base de dados sobre corridas
resultado_corrida =pd.read_csv("results.csv",sep =",") # base de dados sobre os resultados das Corridas
temporada = pd.read_csv("seasons.csv",sep=";") # base de dados sobre os resultados das temporadas
sprint_resultado =pd.read_csv("sprint_results.csv",sep=",") #
status = pd.read_csv("status.csv",sep=",")  #
```
